## Data Distance

```python
import pandas as pd
import numpy as np

from pathlib import Path
```

```python
OUTPUT_FOLDER = Path("output")
data_df = pd.read_csv(OUTPUT_FOLDER / "data.csv").iloc[:, 2:]
data_df
```

```python
from sklearn.metrics.pairwise import euclidean_distances

test = [
    [
      62.6,
      114.0,
      45.0,
      4419.0,
      0.69,
      0.31,
      0.01,
      0.67,
      0.33,
      0.3,
      0.5,
      500000.0,
      40079.0,
      5.25,
      2.75,
      3.0
    ]
]

X = [[*row.to_numpy()]  for i, row in data_df.iloc[:,4:-4].iterrows()]
distance = sorted(list(enumerate(list([value[0] for value in euclidean_distances(X, test)]))), key=lambda x: x[1])


for index, dist_value in distance[:3]:
    print(data_df.iloc[index, :])
```
